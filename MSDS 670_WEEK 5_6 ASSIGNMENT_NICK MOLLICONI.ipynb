### STEP 1 of Analysis: Sum of Facility Type

```python
%matplotlib inline
#get value counts
counts = HEALTHCARE['FAC_FDR'].value_counts()
# Plot the value counts
ax=counts.plot(kind='bar', title='Sum of Facility Type')
# annotate
ax.bar_label(ax.containers[0], label_type='edge')
```

### The sum of Facility Type reflects that Home Health has the highest volume within this dataset.
